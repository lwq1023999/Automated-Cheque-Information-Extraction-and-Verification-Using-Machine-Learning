## YOLOv8 Model Performance Analysis

The table below summarizes the performance metrics of different YOLOv8 models based on their image size, mean Average Precision (mAPval), training time, and computational complexity measured in GFLOPs (B). These models have been evaluated using a common image size of 640 pixels.

| Model   | Image Size (pixel) | mAPval | Time (hours) | GFLOPs (B) |
|---------|---------------------|--------|--------------|------------|
| YOLOv8n | 640                 | 0.693  | 0.063        | 8.1        |
| YOLOv8s | 640                 | 0.703  | 0.116        | 28.4       |
| YOLOv8m | 640                 | 0.711  | 0.968        | 78.7       |
| YOLOv8l | 640                 | 0.708  | 2.29         | 164.8      |
| YOLOv8x | 640                 | X      | X            | X          |

**Model Details:**
- **Image Size (pixel):** The input size of the image used during training and inference.
- **mAPval:** Mean Average Precision on the validation set.
- **Time (hours):** Training time required to reach the reported performance.
- **GFLOPs (B):** Billion FLOating-point Operations per Second, indicating computational complexity.

The table shows that larger YOLOv8 models (e.g., YOLOv8l) achieve higher mAPval but require longer training time and higher computational resources. The performance metrics provide insights into the trade-offs between accuracy, training time, and computational efficiency for different YOLOv8 models.


```python
import matplotlib.pyplot as plt

# Model names and corresponding mAPval scores
models = ['YOLOv8n', 'YOLOv8s', 'YOLOv8m', 'YOLOv8l', 'YOLOv8x']
mAPvals = [0.693, 0.703, 0.711, 0.708, 0.0]  

# Create a bar chart to visualize mAPval scores
plt.figure(figsize=(10, 6))
plt.bar(models, mAPvals, color='skyblue')
plt.title('Mean Average Precision (mAPval) of YOLOv8 Models')
plt.xlabel('Model')
plt.ylabel('mAPval')
plt.ylim(0.6, 0.8)  # Set y-axis limits based on the range of mAPval values
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of paths to CSV files for each YOLOv8 model training run
csv_file_paths = [
    'C:/Users/Admin/Documents/ultralytics/runs/detect/train/results.csv',
    'C:/Users/Admin/Documents/ultralytics/runs/detect/train2/results.csv',
    'C:/Users/Admin/Documents/ultralytics/runs/detect/train3/results.csv',
    'C:/Users/Admin/Documents/ultralytics/runs/detect/train4/results.csv',
    'C:/Users/Admin/Documents/ultralytics/runs/detect/train5/results.csv'
    # Add more paths for additional YOLOv8 model training runs if needed
]

# YOLOv8 model names corresponding to each CSV file path
yolov8_models = ['n', 's', 'm', 'l', 'x']

# Initialize lists to store data for each YOLOv8 model training run
all_epochs = []
all_mAP50_values = []

# Loop through each CSV file path
for csv_file_path in csv_file_paths:
    # Load the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)
    
    # Strip leading and trailing spaces from column names
    df.columns = df.columns.str.strip()
    
    # Check if required columns exist in the DataFrame after stripping spaces
    if all(col in df.columns for col in ['epoch', 'metrics/mAP50(B)']):
        # Extract epoch numbers and mAP50 values
        epochs = df['epoch']
        mAP50_values = df['metrics/mAP50(B)']
        
        # Append epoch numbers and mAP50 values to the lists
        all_epochs.append(epochs)
        all_mAP50_values.append(mAP50_values)
    else:
        print(f"Required columns not found in DataFrame for file: {csv_file_path}")

# Create a stacked line plot for mAP50 values across epochs for each YOLOv8 model training run
plt.figure(figsize=(12, 8))

for epochs, mAP50_values, model_name in zip(all_epochs, all_mAP50_values, yolov8_models):
    plt.plot(epochs, mAP50_values, marker='o', linestyle='-', label=f'YOLOv8 {model_name}')

# Add labels and title
plt.title('mAP50(B) for YOLOv8 Models Across Training Runs')
plt.xlabel('Epoch')
plt.ylabel('mAP50(B)')
plt.grid(True)
plt.legend()

# Display the stacked line plot
plt.show()
```
