## OCR Process Using Pytesseract

### Setting TESSDATA_PREFIX Environment Variable

We first need to set the `TESSDATA_PREFIX` environment variable to specify the directory where Tesseract-OCR's language data files are located. This is necessary for Tesseract to find the required language data files.

```python
import os

# Set TESSDATA_PREFIX environment variable
os.environ['TESSDATA_PREFIX'] = r'C:\Program Files\Tesseract-OCR\tessdata'


````python
import os
import re
import matplotlib.pyplot as plt

# Set TESSDATA_PREFIX environment variable
os.environ['TESSDATA_PREFIX'] = r'C:\Program Files\Tesseract-OCR\tessdata'

from PIL import Image
import pytesseract
````

````python
# Define path to tesseract.exe
path_to_tesseract = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Define path to image
path_to_image = r'C:\Users\Admin\Documents\ultralytics\datasets\chequeinfo-2\test\images\0c430b7d-244_jpg.rf.67633b1307f8700fd5958c01d7e9dedc.jpg'

# Point tesseract_cmd to tesseract.exe
pytesseract.pytesseract.tesseract_cmd = path_to_tesseract

# Open image with PIL
img = Image.open(path_to_image)

# Display the image
plt.imshow(img)
plt.axis('off')  # Turn off axis
plt.show()

# Extract text from image
text = pytesseract.image_to_string(img)

# Use regular expression to separate words and numbers
words = re.findall(r'\b[a-zA-Z]+\b', text)
numbers = re.findall(r'\d+', text)

# Join the extracted words and numbers into single strings
words_str = ' '.join(words)
numbers_str = ''.join(numbers)

print("Extracted words:", words_str)
print("Extracted numbers:", numbers_str)
````
